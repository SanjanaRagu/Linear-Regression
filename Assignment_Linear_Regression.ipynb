#### Extracting New Varivables from the old one

```python
# Creating new variables 
df['carLW'] = df.carlength/df.carwidth
df['carWH'] = df.carwidth/df.carheight
df['PW'] = df.horsepower/df.curbweight
df['HCmpg'] = df.highwaympg/df.citympg
df.drop(['carlength','carwidth','carheight','highwaympg','citympg'],axis=1,inplace=True)
```

#### Dummy Variables

```python
#Dummy variables are created from categorical variables.
#columns with 2 categories can be converted to 0 and 1.
dummy_dict ={'fueltype':{'gas':0,'diesel':1},'aspiration':{'turbo':0,'std':1},'doornumber':{'two':0,'four':1},'enginelocation':{'rear':0,'front':1} }
df.replace(dummy_dict, inplace =True)
```
